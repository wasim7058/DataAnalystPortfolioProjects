```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.columns
```

```python
sns.histplot(data['Government expenditure on education, total (% of government expenditure)'], kde=True)
plt.title('Distribution of Government Expenditure on Education (% of Government Expenditure)')
plt.show()
```

```python
sns.histplot(data['Government expenditure on education, total (% of GDP)'], kde=True)
plt.title('Distribution of Government Expenditure on Education (% of GDP)')
plt.show()
```

```python
sns.boxplot(x='Year', y='Government expenditure on education, total (% of GDP)', data=data)
plt.title('Box Plot of Government Expenditure on Education (% of GDP) by Year')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Country Name', data=data)
plt.xticks(rotation=90)
plt.title('Count of Observations by Country')
plt.show()
```
